```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Part 1 - Plot delay probabilities

```python
delay_prob = pd.read_csv('data/processed/delay-probabilities.csv')

x_axis = delay_prob['AIRLINES'][0:4]

y_axis = delay_prob['delay_prob'][0:4]
plt.bar(x_axis,y_axis,align = 'center')
```

## Part 2 - Plot daily mean airport arrival delays

```python
daily_mean_delay = pd.read_csv('data/processed/daily-mean-delays.csv')
x = daily_mean_delay['FL_DATE'][0:365]
y = daily_mean_delay['ARR_DELAY'][0:365]
plt.figure(figsize=(12,7))
plt.plot(x,y)
plt.xlabel('flight date')
plt.ylabel('delaytime')
plt.yticks(np.arange(min(y), max(y)+1, 5.0))
plt.xticks(np.arange(0, 366, 10),rotation=90)
plt.title('DelayTime vs Flight date at ATL')

```

# Part 3 - Plot histogram of daily mean arrival delays

```python
# YOUR CODE HER\

x_axis = daily_mean_delay['ARR_DELAY'][0:365]; # since this dataframe is small, I checked to find out the the airport ATL corresponds to the first 364 rows in the DEST columns
plt.figure(figsize=(12,7));
plt.hist(x_axis,bins=200);
plt.xlabel('daily_mean_delay')
plt.ylabel('counts')
plt.title('mean_delay vs. counts')
plt.tight_layout()
```
